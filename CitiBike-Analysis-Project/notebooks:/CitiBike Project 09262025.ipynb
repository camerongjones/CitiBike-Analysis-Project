```python
#Hello! Here is my code to visualize some CitiBike rider data! Me and my friends ride citibikes often to get to school and home from work,
#so I thought it would be fun to dive into the CitiBike data on the Lyft website. These csvs are all from SQL queries I made,
#which I will add in a different file if you guys want to replicate this on a different month's data. (This data is all from August 2025.)

#Start by importing our libraries of course.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Creating dfs from each of my SQL queries
by_hour = pd.read_csv("hour_tripdata.csv")
by_day = pd.read_csv("weekday_tripdata.csv")
top_stations = pd.read_csv("top10station_tripdata.csv")
duration_by_type = pd.read_csv("ridertype_tripdata.csv")
```

```python
#Here I am creating a bar chart to show the number of rides in each hour of the day.

plt.figure(figsize=(8,5))
sns.barplot(x="hour", y="trip_count", data=by_hour, color="darkblue")
plt.title("CitiBike Trips by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Trips")
plt.show()
```

```python
#Here I am making a plot for how many rides are taken on each day of the week. 

#First I map the actual days of the week to the integers I have in my dataset for easier comprehension. 
weekday_map = {
    0: "Sunday", 1: "Monday", 2: "Tuesday", 3: "Wednesday",
    4: "Thursday", 5: "Friday", 6: "Saturday"
}
by_day["weekday"] = by_day["weekday"].map(weekday_map)
```

```python
by_day.head()
```

```python
#Now I make the bar chart
plt.figure(figsize=(8,5))
sns.barplot(x="weekday", y="trip_count", data=by_day, color="green",
            order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
plt.title("CitiBike Trips by Weekday")
plt.xlabel("Day of Week")
plt.ylabel("Trips")
plt.show()
```

```python
#Here I am making another barplot to show the top 10 stations riders typically start from, 
#and how many trips started from each of these stations.

plt.figure(figsize=(10,6))
sns.barplot(y="start_station_name", x="trip_count", data=top_stations, color="darkorange")
plt.title("Top 10 CitiBike Start Stations")
plt.xlabel("Trips")
plt.ylabel("Station")
plt.show()
```

```python
#Lastly I am plotting the average duration of a ride based on if the user is a member or a casual user. 

plt.figure(figsize=(6,4))
sns.barplot(x="member_casual", y="avg_duration_minutes", data=duration_by_type, palette="rocket")
plt.title("Average Trip Duration by Rider Type")
plt.xlabel("Rider Type")
plt.ylabel("Avg Duration in Minutes")
plt.show()
```
